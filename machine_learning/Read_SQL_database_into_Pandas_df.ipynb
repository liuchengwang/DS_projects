# Question 1a: Load the data in the SQLite database into a **Pandas** dataframe. The dataframe should be indexed by unique keywords (48 rows), and the columns should correspond to the different weeks (261 columns). 

```python
import pandas as pd
import sqlite3
!ls
```

```python
# load the data from the SQLite database into a Pandas dataframe
with sqlite3.connect('denim_search_data.db') as con:
    df=pd.read_sql("select * from search_vol",con,index_col='keyword',parse_dates=['w0','w1'])
```

```python
print(df.shape)
print(df.dtypes)
print(df.head())
print(df.describe())
```

```python
# rename the column, change the tpye of the column
df=df.rename(columns={'w0':'week_begin_date'})
df['volume']=df['volume'].astype('int64')
print(df.dtypes)
```

```python
# creat a pivot table to answer the question
df2=pd.pivot_table(df,values='volume',index=['keyword'],columns=['week_begin_date'])
```

# Question 1b: Based on the given data, which three denim styles do you believe will be the top performers of 2018? Which three will be the bottom performers?

```python
# For the performers of 2018, I took the data in the week from 2017-12-31 to 2018-01-06 into account.
df3=df[df['week_begin_date']>=pd.datetime(2017,12,31)]
df_year2018=pd.pivot_table(df3,values='volume',index=['keyword'],columns=['week_begin_date'])
```

```python
print(df_year2018.shape)
print(df_year2018.columns)
```

```python
# top/bottom 3 performers of 2018
df_year2018['sum_2018']=df_year2018.sum(axis=1)
print('=== top 3 performers of 2018 ===')
print(df_year2018[['sum_2018']].nlargest(3,'sum_2018'))
print
print('=== bottom 3 performers of 2018 ===')
print(df_year2018[['sum_2018']].nsmallest(3,'sum_2018'))
```
